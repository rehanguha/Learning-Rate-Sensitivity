```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter
```

```python
df = pd.read_csv("output/CIFAR10_LR_COPY.csv")
```

```python
df
```

```python
df.optimizer.value_counts()
```

```python

sns.lineplot(data=df[df['optimizer'].isin(['SGD',
                                          'Adam',
                                          'Adadelta',
                                          'RMSprop',
                                          'Nadam',
                                          'Adagrad',
                                          'Adamax'])], 
                x='lr', 
                y='test_accuracy', hue="optimizer")
```

```python

sns.lineplot(data=df[df['optimizer'].isin(['Adagrad',
                                          'Adadelta', 'SGD'])], 
                x='lr', 
                y='test_accuracy', hue="optimizer")
```

```python

sns.lineplot(data=df[df['optimizer'].isin(['RMSprop',
                                          'Adam',
                                          'Nadam',''])], 
                x='lr', 
                y='accuracy', hue="optimizer")
```

```python

sns.lineplot(data=df[df['optimizer'].isin(['SGD',
                                          'Adadelta',
                                          'Adagrad',
                                          'Adamax'])], 
                x='lr', 
                y='test_accuracy', hue="optimizer")
plt.xlabel('Learning Rate')
plt.ylabel('Test Accuracy')
```

```python

sns.lineplot(data=df[df['optimizer'].isin(['SGD',
                                          'Adadelta',
                                          'Adagrad',
                                          'Adamax'])], 
                x='lr', 
                y='accuracy', hue="optimizer")
plt.xlabel('Learning Rate')
plt.ylabel('Train Accuracy')
```

```python
fig = plt.figure(figsize=(10,600))

fig, ax = plt.subplots(ncols=2, 
                       sharex=True, 
                       sharey=False, 
                       constrained_layout=True,
                       figsize=(15,5))


sns.lineplot(data=df[df['optimizer'].isin(['Adagrad',
                                          'Adadelta','SGD'])], 
                x='lr', 
                y='accuracy', hue="optimizer", ax=ax[0])
ax[0].set_ylabel("Train Accuracy")
ax[0].set_xlabel("Learning Rate")
ax[0].legend(title='Optimizers')
ax[0].title.set_text("Train Accuracy vs. LR")
ax[0].xaxis.set_major_formatter(FormatStrFormatter('%.3f'))
plt.xticks(np.arange(0, 0.101, 0.01))


sns.lineplot(data=df[df['optimizer'].isin(['Adagrad',
                                          'Adadelta', 'SGD'])], 
                x='lr', 
                y='test_accuracy', hue="optimizer", ax=ax[1])
ax[1].set_ylabel("Test Accuracy")
ax[1].set_xlabel("Learning Rate")
ax[1].legend(title='Optimizers')
ax[1].title.set_text("Test Accuracy vs. LR")
```
